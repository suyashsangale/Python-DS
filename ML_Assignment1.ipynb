```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   
import seaborn as sns
%matplotlib inline
```

```python
df["Used_in_years"]=2019-df["Year"]
df=df.drop(["Year"],1)
```

```python
sns.pairplot(df)
```

```python
import numpy as np
numerical = df.select_dtypes(include=np.number)
categorical=df.select_dtypes(exclude=np.number)
```

```python
numerical=numerical.drop(["Owner"],1)
```

```python
numerical.head(10)
```

```python
df_n = np.log(numerical)
```

```python
df_n.head(20)
```

```python
sns.pairplot(df_n)
```

```python
df=df.drop(["Selling_Price","Present_Price","Kms_Driven","Used_in_years"],1)
```

```python
df_new=pd.concat([df,df_n],1)
```

```python
df_new.head(10)
```

```python
X=df_new.drop(["Selling_Price"],1)
```

```python
X=pd.get_dummies(X)
```

```python
y=df_new["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.33, random_state=42)
```

```python
X_train.dtypes
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
sfs=SFS(RandomForestRegressor(n_jobs=4),k_features=10,forward=True,floating=False,verbose=1,scoring='r2',cv=3)
```

```python
sfs=sfs.fit(np.array(X_train.fillna(0)),y_train)
select_feat_forward= X_train.columns[list(sfs.k_feature_idx_)]
select_feat_forward
```
